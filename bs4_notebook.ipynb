```python
URL = "https://bocongan.gov.vn/khung-bo/to-chuc-khung-bo/to-chuc-khung-bo-viet-tan-8.html?Page=1"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

urls = []
max_page = None
try:
    max_page = len(soup.find_all(class_="paging")[0].find_all())
    for i in range(1,max_page+1):
        url = f"https://bocongan.gov.vn/khung-bo/to-chuc-khung-bo/to-chuc-khung-bo-viet-tan-8.html?Page={i}"
        urls.append(url)
except:
    urls.append(URL)


urls
```

```python
def clean_html_string(input_text):
    clean_html = re.sub(r'<.*?>', '', input_text)
    # Remove tabs and newlines
    clean_text = clean_html.replace('\t', '').replace('\n', '')
    return clean_text
```

```python
_member_pages = []
member_vt_data_list = []
vt_metadata = {
            "type": None,
            "source": None,
            "watchlist": None,
            "doc_local_path": None,
        }
for url_page_vt in urls:
    _page = requests.get(url_page_vt)
    _soup = BeautifulSoup(_page.content, "html.parser")
    for tag in _soup.find_all(class_="imgthanhvien"):
        member = tag.find_all()[1]
        member_link = member.get('href')
        _member_pages.append('https://bocongan.gov.vn/'+member_link)

for page in _member_pages:
    # subpage = requests.get(page)
    subpage = requests.get(page)
    sub_soup = BeautifulSoup(subpage.content, "html.parser")
    tmp_sub_soup = sub_soup
    detail_data = tmp_sub_soup.find_all(class_="table-detail")
    detail_data = clean_html_string(str(detail_data[0]))
    _metadata = vt_metadata.copy()
    _metadata['type'] = 'individual'
    _metadata['source'] = str(page)
    _metadata['watchlist'] = 'sanctions'
    member_vt_data_list.append((detail_data, _metadata))
    


```
